We see that the Maclaurin formula for the sine function is given by $$\sum_{k = 0}^{\infty} \frac{-1^{k} x^{2k} }{(2k)!}.$$ We will only calculate a finite expansion $$\sum_{k = 0}^{n} \frac{-1^{k} x^{2k} }{(2k)!},$$ for some positive integer $n$.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create data to plot
x = np.linspace(-10, 10, 100)
y = np.sin(x)

# Plot the data
plt.plot(x, y, color='red')
```

# Using python functions and numpy.math for factorial
We see here that the upper limit for $n$ is 171. Any higher values for $n$ causes an overflow error - mainly one that says the number $x^{n}$ for $n \geq 172$ has an integer part too large to convert to a float. We use numpy to calculate the factorial, and python's default exponentiation operator.

Below there are two graphs showing the values of clear divergence.

```python
# Initialization
x = np.linspace(-50, 50, 1000)
y = np.sin(x)

# Plot the data of regular sine
plt.plot(x, y, color='red')

# Taylor expansion
colorList = ['black', 'orange', 'blue', 'purple', 'green']
n = 172
y_1 = 0
for i in range(0, n):
    if (i % 4 == 1):
        y_1 = y_1 + ((x**i) / np.math.factorial(i))
    elif (i % 4 == 3):
        y_1 = y_1 - ((x**i) / np.math.factorial(i))
plot1 = plt.plot(x, y_1, color=colorList[(4 // 2) % 5])
plt.xlim(30.0,40.0)
plt.ylim(-2.0,2.0)
```

```python
# The other end
y = np.sin(x)
plt.plot(x, y, color='red')

plot2 = plt.plot(x, y_1, color=colorList[(4 // 2) % 5])
plt.xlim(-40.0,-30.0)
plt.ylim(-2.0,2.0)
```

```python
# Overall picture:
y = np.sin(x)
plt.plot(x, y, color='red')

plot2 = plt.plot(x, y_1, color=colorList[(4 // 2) % 5])
plt.xlim(-50,50)
plt.ylim(-2,2)
```

# Using numpy for power and numpy.math for factorial
Replacing the python exponentiation operator with numpy's power function, setting the data type to be an explicit float64, and initializing the value of $y_1$ to be a float64 does NOT increase the maximum value of $n$ the loop can calculate without causing an overflow error:

```python
x = np.linspace(-50, 50, 1000)
y = np.sin(x)

# Plot the data of regular sine
plt.plot(x, y, color='red')

# Taylor expansion
colorList = ['black', 'orange', 'blue', 'purple', 'green']
n = 171
y_1 = np.float64(0.0)
for i in range(0, n):
    if (i % 4 == 1):
        y_1 = y_1 + (np.power(x,i,dtype=np.float64) / np.math.factorial(i))
    elif (i % 4 == 3):
        y_1 = y_1 - (np.power(x,i,dtype=np.float64) / np.math.factorial(i))
plt.plot(x, y_1, color=colorList[(4 // 2) % 5])
plt.xlim(30,40)
plt.ylim(-2,2)
```

```python
# Regular sine for reference
y = np.sin(x)
plt.plot(x, y, color='red')

plot2 = plt.plot(x, y_1, color=colorList[(4 // 2) % 5])
plt.xlim(-40,-30)
plt.ylim(-2,2)
```

```python
# Overall picture:
y = np.sin(x)
plt.plot(x, y, color='red')

plot2 = plt.plot(x, y_1, color=colorList[(4 // 2) % 5])
plt.xlim(-50,50)
plt.ylim(-2,2)
```
